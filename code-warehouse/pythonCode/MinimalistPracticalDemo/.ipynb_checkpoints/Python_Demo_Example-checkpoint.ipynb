```python
from math import ceil

def chunk(lst, size):
    return list(
     map(lambda x: lst[x * size:x * size + size],
        list(range(0, ceil(len(lst) / size)))))

chunk([1,2,3,4,5],2)
```

#### å±•å¼€åˆ—è¡¨
è¯¥æ–¹æ³•å°†é€šè¿‡é€’å½’çš„æ–¹å¼å°†åˆ—è¡¨çš„åµŒå¥—å±•å¼€ä¸ºå•ä¸ªåˆ—è¡¨ã€‚

```python
def spread(arg):
    ret = []
    for i in arg:
        if isinstance(i, list):
            ret.extend(i)
        else:
            ret.append(i)
    return ret

def deep_flatten(lst):
    result = []
    result.extend(spread(list(map(lambda x: deep_flatten(x) if type(x) == list else x, lst))))
    return result
deep_flatten([1, [2], [[3], 4], 5])
```

#### åˆ—è¡¨çš„å·®
è¿”å›ç¬¬ä¸€ä¸ªåˆ—è¡¨ä¸åœ¨ç¬¬äºŒä¸ªåˆ—è¡¨ä¸­çš„å…ƒç´ ä¸è¿”å›ç¬¬äºŒä¸ªåˆ—è¡¨ä¸åœ¨ç¬¬ä¸€ä¸ªåˆ—è¡¨ä¸­çš„å…ƒç´ 

```python
def difference(a, b):
    set_a = set(a)
    set_b = set(b)
    comparison = set_a.difference(set_b)
    comparison2 = set_b.difference(set_a)
    return list(comparison),list(comparison2)

difference([1,2,3], [1,2,4])
```

#### é€šè¿‡å‡½æ•°å–å·®
å¦‚ä¸‹æ–¹æ³•é¦–å…ˆä¼šåº”ç”¨ä¸€ä¸ªç»™å®šçš„å‡½æ•°ï¼Œç„¶åå†è¿”å›åº”ç”¨å‡½æ•°åç»“æœæœ‰å·®åˆ«çš„åˆ—è¡¨å…ƒç´ ã€‚

```python
def difference_by(a, b, fn):
    b = set(map(fn, b))
    return [item for item in a if fn(item) not in b]

from math import floor
print(difference_by([2.1, 1.2], [2.3, 3.4],floor))
print(difference_by([{ 'x': 2 }, { 'x': 1 }], [{ 'x': 1 }], lambda v : v['x']))
```

#### åˆå¹¶ä¸¤ä¸ªå­—å…¸
ä¸‹é¢çš„æ–¹æ³•å°†ç”¨äºåˆå¹¶ä¸¤ä¸ªå­—å…¸ã€‚

```python
def merge_two_dicts(a, b):
    c = a.copy()  # make a copy of a
    c.update(b)  # modify keys and values of a with the ones from b
    return c

a = { 'x': 1, 'y': 2}
b = { 'y': 3, 'z': 4}
print(merge_two_dicts(a, b))
```

```python
def merge_dictionaries(a, b):
    return {**a, **b}

a = { 'x': 1, 'y': 2}
b = { 'y': 3, 'z': 4}
print(merge_dictionaries(a, b))
```

#### å…ƒç´ é¢‘ç‡

```python
def most_frequent(list):
    return max(set(list), key = list.count)

list = [1,2,1,2,3,2,1,4,2]
most_frequent(list) 
```

#### å›æ–‡åºåˆ—
ä»¥ä¸‹æ–¹æ³•ä¼šæ£€æŸ¥ç»™å®šçš„å­—ç¬¦ä¸²æ˜¯ä¸æ˜¯å›æ–‡åºåˆ—ï¼Œå®ƒé¦–å…ˆä¼šæŠŠæ‰€æœ‰å­—æ¯è½¬åŒ–ä¸ºå°å†™ï¼Œå¹¶ç§»é™¤éè‹±æ–‡å­—æ¯ç¬¦å·ã€‚æœ€åï¼Œå®ƒä¼šå¯¹æ¯”å­—ç¬¦ä¸²ä¸åå‘å­—ç¬¦ä¸²æ˜¯å¦ç›¸ç­‰ï¼Œç›¸ç­‰åˆ™è¡¨ç¤ºä¸ºå›æ–‡åºåˆ—ã€‚

```python
def palindrome(string):
    from re import sub
    s = sub('[\W_]', '', string.lower())
    return s == s[::-1]

palindrome('taco cat')
```
