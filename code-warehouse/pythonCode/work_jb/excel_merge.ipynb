```python
import sys,os,shutil,types,time,json,traceback,datetime
import xlrd
from openpyxl import load_workbook
from openpyxl import Workbook
from openpyxl.writer.excel import ExcelWriter


def getFiles(path,keyword):
    files=[x for x in os.listdir(path) if os.path.isfile(os.path.join(path,x)) and os.path.splitext(x)[1]=='.xls']
    filterFiles=[]
    for f in files:
        if(keyword==None or f.find(keyword)>=0):
            filterFiles.append(os.path.join(path,f))
    return filterFiles
def isNoneOrWhitespace(str):
    if(str==None):
        return True
    str=str.strip()
    if(len(str)==0):
        return True
    return False
def hanldeFile(excelFile,ws):
    if ws.max_row<2:
        ws.cell(row=1,column=1).value="F_YEAR"
        ws.cell(row=1,column=2).value="F_0001"
        ws.cell(row=2,column=1).value="年度"
        ws.cell(row=2,column=2).value="国家"
    rb=xlrd.open_workbook(excelFile)
    sh=rb.sheet_by_index(0)
    indexName=None
    indexCol=None
    for r in range(1,sh.nrows):
        if(isNoneOrWhitespace(sh.cell_value(r,1))):
            break
        if(not isNoneOrWhitespace(sh.cell_value(r, 0))):
            indexName=sh.cell_value(r,0)
            indexCol=None
            for wc in range(3,ws.max_column+1):
                if(ws.cell(row=2,column=wc).value==indexName):
                    indexCol=wc
            if(indexCol==None):
                indexCol=ws.max_column+1
                ws.cell(row=2,column=indexCol).value=indexName
            # else:
            #     print("Waring:",excelFile,indexName)
        country=sh.cell_value(r, 1)
        for c in range(2,sh.ncols):
            year=sh.cell_value(0,c)
            val='%s'%(sh.cell_value(r,c))
            wr=None
            for ir in range(1,ws.max_row+1):
                if(ws.cell(row=ir,column=1).value==year and ws.cell(row=ir,column=2).value==country):
                    wr=ir
                    # if(ws.cell(row=wr,column=indexCol).value!=None):
                    #     print("Error",excelFile,indexName,indexCol,ws.cell(row=wr,column=indexCol).value,val)
            if(wr==None):
                wr=ws.max_row+1
                ws.cell(row=wr,column=1).value=year
                ws.cell(row=wr,column=2).value=country
            ws.cell(row=wr,column=indexCol).value=val
def handleFiles(path,savefile):
    files=getFiles(path,None)
    wb=Workbook()
    ws=wb.worksheets[0]
    for f in files:
        print(f)
        hanldeFile(f,ws)
    wb.save(savefile)
# def Test():
#     wb=Workbook()
#     ws=wb.worksheets[0]
#     hanldeFile("C:\\Users\\ding\\Desktop\\python\\data卫生\\国际数据-世界经济发展数据库-卫生-1.xls",ws)
#     wb.save("C:\\Users\\ding\\Desktop\\python\\data卫生.xlsx")


# Test()
sourcePath=r'C:\\Users\\ding\\Desktop\\python\\data外债'
savePath=r'C:\\Users\\ding\\Desktop\\python\\data外债\\外债.xlsx'
handleFiles(sourcePath,savePath)
```

```python
import pandas as pd
import numpy as np
import os
def handle_data(open_files,save_file):
    os.chdir(open_files)
    file_chdir = os.getcwd()
    filecsv_list = []
    for root,dirs,files in os.walk(file_chdir):
        for file in files:
            if os.path.splitext(file)[1] == '.xls':
                filecsv_list.append(file)
    df = pd.DataFrame()
    for excel in filecsv_list:
        df = df.append(pd.read_excel(excel,header = 0,sep=None,encoding='gb18030'))
    df = df.rename(columns={'Unnamed: 0':'指标','Unnamed: 1':'国家'})
    # 根据上行的数据填充
    df['指标']=df['指标'].fillna(method='ffill')
    # melt进行行转列
    df2 = pd.melt(df, id_vars=["指标",'国家'],var_name="年份",value_name="数据")
    # pivot_table 转为想要的格式,dropna=False不忽略空值
    df3 = pd.pivot_table(df2,index=['国家','年份'],columns=['指标'],values=['数据'])
    df3.to_excel(save_file)
file_addr = 'D:\Desktop\pythonData\一带一路'
save_addr = '农产品加工.xlsx'
handle_data(file_addr,save_addr)

#value_vars=["2015", "2016", "2017",'2018','2019']
# id_vars不需要转换的列
# value_vars=["2015"...] 需要转换的列，如果都需要就不用写
# var_name和value_name是自定义设置对应的列名。
# col_level :如果列是MultiIndex，则使用此级别
```
